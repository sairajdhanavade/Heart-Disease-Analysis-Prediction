```python
x=sns.countplot(data=df,x="Heart Disease")
plt.title("distribution of patient Having Heart Disease or Not")
plt.bar_label(x.containers[0])
plt.show()
```

## Insights
1. 150 patient are not have the Heart Disease
2. 120 patient are have the Heart Disease

```python
x=sns.countplot(data=df,x=pd.cut(df.Age,bins=[29,47,62,77],labels=["29-47","48-62","63-77"]))
plt.bar_label(x.containers[0])
plt.title("Number of patient in different Age groups")
plt.show()
```

```python
x=sns.countplot(data=df,x=pd.cut(df.Age,bins=[29,47,62,77],labels=["29-47","48-62","63-77"]),hue="Heart Disease")
plt.bar_label(x.containers[0])
plt.bar_label(x.containers[1])
plt.title("Number of patient Having Heart Disease or Not in different Age groups")
plt.show()
```

## Insights
1. 149 patient are having age between 48-62, out of that 76 having heart disease and 73 are do not have heart disease
2. 66 patient are having age between 29-47, out of that 18 having heart disease and 48 are do not have heart disease
3. 54 patient are having age between 63-77, out of that 26 having heart disease and 28 are do not have heart disease

```python
x=sns.countplot(data=df,x="Sex")
plt.bar_label(x.containers[0])
x.set_xticklabels(["Female","Male"])
plt.title("distribution of Male and Female Patient")
plt.show()
```

```python
x=sns.countplot(data=df,x="Sex",hue="Heart Disease")
plt.bar_label(x.containers[0])
plt.bar_label(x.containers[1])
x.set_xticklabels(["Female","Male"])
plt.title("distribution of Male and Female Patient having Heart Disease or Not")
plt.show()
```

## Insights

2. out of 87 female patient 20 patient suffering from Heart Disease

```python
plt.figure(figsize=(10,5))
x=sns.countplot(data=df,x="Chest pain type")
plt.bar_label(x.containers[0])
plt.title("Number of paitent a/c to Chest pain type")
x.set_xticklabels(["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
x=sns.countplot(data=df,x="Chest pain type",hue="Heart Disease")
plt.bar_label(x.containers[0])
plt.bar_label(x.containers[1])
plt.title("Chest pain type VS Heart Disease")
x.set_xticklabels(["Typical angina","Atypical angina","Non-anginal pain","Asymptomatic"])
plt.show()
```

## Insights
1. out of 129 paitent of Asymptomatic chest pain type 91 suffering from Heart Disease

```python
plt.figure(figsize=(10,5))
x=sns.countplot(data=df,x="EKG results")
plt.bar_label(x.containers[0])
plt.title("Number of paitent a/c to electrocardiogram results")
x.set_xticklabels(["Normal","ST-T wave abnormality","Left ventricular hypertrophy"])
plt.show()
```

```python
x=sns.countplot(data=df,x="Exercise angina")
plt.title("Distribution of patient which Exercise-induced angina")
plt.bar_label(x.containers[0])
x.set_xticklabels(["No","Yes"])
plt.show()
```

```python
sns.countplot(data=df,x="FBS over 120",palette={"b","g"})
plt.title("Distribution of paitent on the basis of Fasting blood sugar")
plt.legend(["False","True"])
plt.show()
```
